# 实验四的绘图

绘制的是三条线的的图，纵坐标是 $\frac I I_0$ 纵坐标是 $f/f_0$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

# 设置中文字体和字体大小  如果显示乱码就是没有安装对应的字体，此时可是下载对应的字体也可以更换下面的font.family的值
plt.rcParams['font.family'] = 'Noto Sans CJK JP'  # 设置全局中文字体
plt.rcParams['font.size'] = 12          # 设置全局字体大小
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 数据点
x = np.array([0.234, 0.467, 0.822, 0.934, 0.953, 0.981, 1.028, 1.056, 1.093, 1.214, 1.401, 1.681, 1.868])

# 黄色三角形曲线
y1 = np.array([0.0343, 0.0831, 0.333, 0.716, 0.822, 0.96, 0.921, 0.786, 0.619, 0.342, 0.202, 0.129, 0.105
])

# 蓝色正方形曲线
y2 = np.array([0.0525, 0.126, 0.474, 0.845, 0.909, 0.983, 0.971, 0.895, 0.774, 0.482, 0.299, 0.194, 0.159])

# 红色圆形曲线
y3 = np.array([0.2, 0.445, 0.899, 0.985, 0.992, 0.999, 0.999, 0.994, 0.98, 0.909, 0.774, 0.609, 0.531])

# 插值生成平滑曲线
spl1 = make_interp_spline(x, y1, k=3)
spl2 = make_interp_spline(x, y2, k=3)
spl3 = make_interp_spline(x, y3, k=3)

# 生成平滑的x值
x_smooth = np.linspace(0.2, 2, 500)

# 计算平滑曲线的y值
y1_smooth = spl1(x_smooth)
y2_smooth = spl2(x_smooth)
y3_smooth = spl3(x_smooth)

# 绘图
#记得调整label
plt.figure(figsize=(8, 6))
plt.plot(x_smooth, y1_smooth, label='黄色', color='yellow')
plt.plot(x_smooth, y2_smooth, label='蓝色', color='blue')
plt.plot(x_smooth, y3_smooth, label='红色', color='red')

# 标记数据点
plt.scatter(x, y1, color='yellow')
plt.scatter(x, y2, color='blue')
plt.scatter(x, y3, color='red')

# 添加标题和标签（可以单独设置字体属性）
plt.title('graph', fontsize=14, fontweight='bold')  # 标题加大加粗
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)

# 显示图例（设置图例字体属性）
plt.legend(prop={'size': 10}, loc='best')

# 显示图形
plt.grid(True)
plt.tight_layout()  # 自动调整子图参数
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置中文字体和字体大小  如果显示乱码就是没有安装对应的字体，此时可是下载对应的字体也可以更换下面的font.family的值
plt.rcParams['font.family'] = 'Noto Sans CJK JP'  # 设置全局中文字体
plt.rcParams['font.size'] = 12          # 设置全局字体大小
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
# 设置参数
frequency = 50  # 频率50Hz
phase_diff = 12.87  # 相位差xx度
amplitude1 = 20  # 第一个正弦曲线的峰值
amplitude2 = 20 # 第二个正弦曲线的峰值，明显大于第一个或等于

# 创建时间数据
t = np.linspace(-0.001, 0.04, 1000)  # 时间从0到0.1秒，总共500个点

# 创建两个正弦波
y1 = amplitude1 * np.sin(2 * np.pi * frequency * t)
y2 = amplitude2 * np.sin(2 * np.pi * frequency * t + np.radians(phase_diff))  # 加上相位差

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(t, y1, label="1°", color='b')
plt.plot(t, y2, label="2", color='r')
plt.xlabel("时间")
plt.ylabel("幅值")
plt.title("电路相位差示意图")
plt.legend()
plt.grid(True)
plt.show()
```
